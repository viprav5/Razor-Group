### Count of positive ratings

### Parsing seller email

```python
def rmv(x):
    if x[-2:]!='CN':
        return True
    else:
        return False
```

```python
df_1 = df[df['businessaddress'].apply(rmv)]
```

```python
import re
```

```python
df_1['selleremail'] = df['sellerdetails'].apply(lambda x: re.findall(r"[a-z0-9\.\-+]+@[a-z0-9\.\-+]+\.[a-z]+", x))
```

```python
df_1.head()
```

### Parsing phone number

```python
def find_phone_number(text):
    ph_no = re.findall(r"\b\d{10}\b",text)
    return "".join(ph_no)
```

```python
df_1['mobile'] = df_1['sellerdetails'].apply(find_phone_number)
```
